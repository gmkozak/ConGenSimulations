```r
library(cgsim)
```

```r
set.seed(10)
G <- gt_sim(n = 1000, m = 15, a = 5, asim = TRUE, c1 = c(0.01, 0.07, 0.10), c2 = c(0.10, 0.20 , 0.6), generations = 100)
G1 <- G$gen[G$pop == 1, ]
G2 <- G$gen[G$pop == 2, ]
G3 <- G$gen[G$pop == 3, ]
he(G$gen[G$pop == 1, ])
```

```r
mrr <- c(0, 0, 0.5, 0.75, 1, 0.75, 0.5, 0) # male reproduction rates
frr <- c(0, 0, 0.5, 0.75, 1, 0.75, 0.5, 0) # female reproduction rates
msr <- c(rep(0.65, 7), 0) # male survival rates
fsr <- c(rep(0.65, 7), 0) # female survival rates
burn <- 100 # burn-in iterations
nind <- 100 # initial population size
years <- 100 # simulation years
reps <- 10 # replicates
gm <- "ann" # annual growth model
gr <- c(0.045, 0.055) # percentage of annual population change
nvar <- 0.05 # variation to introduce in replicate starting population size
tsr <- c(0.25, 0.25) # transplants are 40% less likely to survive
tr_in <- list(af(G2), af(G2), af(G2)) # list of transplant allele frequencies
ntran <- c(8, 8, 8) # number of transplants
ytran <- c(15:17) # year of augmentation
yeffect <- 2 # number of years transplant survival rates apply
```

```r
E3 <- effect_sim(pop = G1, syears = 1:years, n = nind, n_rep = reps, 
                 growth_model = gm, growth_rate = gr, nvar = nvar,
                 burnin = burn, mrr = mrr, msr = msr, frr = frr, fsr = fsr,
                 
                 trans = tr_in, tsim = c(F, F, F), tsr = tsr,
                 tyears = ytran, ntrans = ntran, trans_year = yeffect,
                 skip_recruit = FALSE)
```

```r
Get the predicted heterzygosity after the translocation (year 18) and after the simuation (100 year)
```

```r
pred <- gd_ret_pred(out = E3, gd = "He") 

print ("He after the translocation")
pred[18,]

print("He after 100 generations")
pred[100,]
```

```r
par(mfrow = c(1, 2), mar = c(4, 4, 0.5, 0.2), oma = c(0, 0, 0, 0))
pop <- E3$n_tracking[, , 2]
pred <- gd_ret_pred(out = E3, gd = "He") 
plot(NA, ylim = c(0, max(pred)), xlim = c(1, years), ylab = expression("Prop. "~GD[0]), xlab = "Year")
mtext("(a)", side = 3, adj = -0.3, line = -1)
xaxt <- seq(1, years + 1, 10)
points(xaxt, pred[xaxt, 1], pch = 19)
segments(xaxt, pred[xaxt, 2], xaxt, pred[xaxt, 3])
plot(NA, ylim = c(0, max(pop)), xlim = c(1, years), ylab = "Pop. Size", xlab = "Year")
mtext("(b)", side = 3, adj = -0.3, line = -1)
for (i in 1:reps) lines(pop[, i], col = "lightgray")
```
